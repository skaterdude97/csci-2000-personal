```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy import misc
from skimage import color


peppers =misc.imread('peppers.png')
shape = peppers.shape
print("the dimensions of the RGB image are", shape[0], "x", shape[1], "x", shape[2])

gray = color.rgb2gray(peppers, )
plt.imshow(gray,cmap=plt.cm.gray)
misc.imsave('peppers-gray.png', gray)

print("For the grayscale version of peppers.png:")
print("Minimum:", gray.min())
print("Maximum:", gray.max())
size= shape[0]*shape[1]
lst = ()
for i in len(gray):
    for j in gray[i]:
        lst.add(j)
lst.sort()
print("Median:", lst[size/2])
```
